```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
mov_data = pd.read_csv('~/Documents/python/MoviesOnStreamingPlatforms_updated.csv', index_col = 0)
```

```python
tv_data = pd.read_csv('~/Documents/python/tv_shows.csv', index_col = 0)
```

```python
mov_data.head(5)
```

```python

tv_data.head(5)
```

```python
#contat Movies and TV Shows Data:
movies_data = pd.concat([mov_data, tv_data], sort =True)

movies_data
```

```python
#Some stats/info on the dataset
movies_data.info()
```

```python
#Names of Columns in Dataset
movies_data.columns
```

```python
#look for duplicates
movies_data.duplicated()
```

```python
#missing values in columns
total = movies_data.isnull().sum().sort_values(ascending=False)
percent = (movies_data.isnull().sum()/movies_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head()
```

```python
#Filling NaN Value - 
#AGE
movies_data.Age = movies_data.Age.fillna('Unknown')

#IMDb
movies_data.IMDb = movies_data.IMDb.fillna(0)

#Rotten Tomatoes
movies_data['Rotten Tomatoes'] = movies_data['Rotten Tomatoes'].fillna(0)

#Country - 
movies_data['Country'] = movies_data.Country.fillna('United States')

#language
movies_data['Language']= movies_data.Language.fillna('Unknown')

#Genres
movies_data['Genres']= movies_data.Genres.fillna('Unknown')



```

```python
#Dropping unwanted coloum
col = ['Directors']
movies_data.drop(col,axis=1, inplace=True)
```

```python
#Create funtion to obtain 50 pencentile of column
def pct50 (column):
    return column.quantile(0.5)
```

```python
movies_data["Runtime"].agg(pct50)
```

```python
#Filling NaN value 'Runtime'-
movies_data['Runtime']= movies_data.Runtime.fillna(92.0)
```

```python
#Replace TYPE values as Movie/Series
movies_data['Type'].replace([0,1],['Movie','Series'],inplace=True)
```

```python
movies_data
```

```python
#missing values in columns
total = movies_data.isnull().sum().sort_values(ascending=False)
percent = (movies_data.isnull().sum()/movies_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head()
```

```python
#Visualization (1)
#languages which are mostly represented in the streaming services of our dataset
languages = movies_data.Language.value_counts().head(20)

plt.figure(figsize=(12,8))
plt.title('Top Movie Languages on Streaming Services')
sns.barplot(x=languages.values, y=languages.index)
plt.show()
```

```python
plat_dist = movies_data[["Netflix", "Prime Video","Hulu","Disney+"]].sum().sort_values(ascending=False).reset_index()
plat_dist = pd.DataFrame(plat_dist.values, columns=['Platform',"No of Movies"])
```

```python
#Visualisation (2)
plt.figure(figsize=(10,7))
dt = plat_dist["No of Movies"]
labels = plat_dist["Platform"]
exp = (0.1,0.1,0.1,0.1)
clr = ("cyan", "green","red","yellow")
wp = {'linewidth': 2, 'edgecolor':"black"}
plt.pie(dt,labels=labels, autopct='%1.1f%%',explode=exp, colors=clr, shadow=True, wedgeprops=wp)
plt.title('Distribution of Movies among the Platforms')
plt.axis('equal')
plt.show()
```

```python
#What Streaming service has the most classic movies?
classic_movies_df = movies_data.loc[movies_data.Year < 1985]
chosen_columns = ['Netflix','Prime Video','Disney+','Hulu']
classic_movies = classic_movies_df[chosen_columns].sum()
plt.figure(figsize=(10,6))
plt.title('Classic Movies Available pre 1985')
sns.barplot(x=classic_movies.values, y=classic_movies.index)
plt.xticks(np.arange(0,2500,250))
plt.show()
```
